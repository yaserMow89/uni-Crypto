```python
import random
from Crypto.Util import number

def generate_rsa_keys(keysize=2048):
    # Generate prime numbers p and q
    p = number.getPrime(keysize // 2)
    q = number.getPrime(keysize // 2)
    
    n = p * q
    phi = (p-1) * (q-1)
    
    # Choose an integer e such that e and phi(n) are coprime
    e = 65537
    
    # Compute d, the mod inverse of e
    d = number.inverse(e, phi)
    
    # Return the public and private keys
    return ((e, n), (d, n))

def rsa_encrypt(public_key, plaintext):
    # Convert plaintext to integers based on UTF-8 encoding
    plaintext_integers = [ord(char) for char in plaintext]
    
    # Encrypt plaintext using the public key
    e, n = public_key
    ciphertext = [pow(char, e, n) for char in plaintext_integers]
    
    return ciphertext

def rsa_decrypt(private_key, ciphertext):
    # Decrypt ciphertext using the private key
    d, n = private_key
    decrypted_integers = [pow(char, d, n) for char in ciphertext]
    
    # Convert integers back to string based on UTF-8 encoding
    plaintext = ''.join(chr(char) for char in decrypted_integers)
    
    return plaintext

# Example usage
public_key, private_key = generate_rsa_keys(2048)

plaintext = "The quick brown fox jumped over the lazy dog"
print(f"Original: {plaintext}")

ciphertext = rsa_encrypt(public_key, plaintext)
print(f"Encrypted: {ciphertext}")

decrypted_text = rsa_decrypt(private_key, ciphertext)
print(f"Decrypted: {decrypted_text}")
```
